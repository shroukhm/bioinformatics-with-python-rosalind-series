## 1. Basic Programming


manually check each character in the DNA string and replace 'T' with 'U' to form the RNA string.

```python
dna = "GATGGAACTTGACTACGTAAATT"
rna = ""

for nucleotide in dna:
    if nucleotide == 'T':
        rna += 'U'
    else:
        rna += nucleotide

print(rna)
```

## 2 . Using Built-in String Function

uses Python's built-in `str.replace()` method to replace all 'T' characters with 'U' in one line.

```python
dna = "GATGGAACTTGACTACGTAAATT"
rna = dna.replace("T", "U")

print(rna)
```

##  3. Using Biopython Library

This solution uses the `Bio.Seq` module from the Biopython library. It calls the `.transcribe()` method, which mimics biological transcription.

```python
!pip install biopython
```

```python
from Bio.Seq import Seq

dna = Seq("GATGGAACTTGACTACGTAAATT")
rna = dna.transcribe()

print(rna)
```
