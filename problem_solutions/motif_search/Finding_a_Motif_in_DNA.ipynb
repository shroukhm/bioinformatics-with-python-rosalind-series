# From Scratch ( basic programming )


```python
def find_motif_manual(s, t):
    positions = []
    len_s = len(s)
    len_t = len(t)

    # Check every possible starting position
    for i in range(len_s - len_t + 1):
        match = True

        # Check if t matches s starting at position i
        for j in range(len_t):
            if s[i + j] != t[j]:
                match = False
                break

        if match:
            positions.append(str(i + 1))  # 1-based position

    return positions

# Example
s = "GATATATGCATATACTT"
t = "ATAT"
result = find_motif_manual(s, t)
print(" ".join(result))
```

# Using Built-in find()


```python
def find_motif_positions(s, t):
    """Find all 1-based starting positions of t in s"""
    positions = []
    start = 0

    while True:
        pos = s.find(t, start)
        if pos == -1:
            break
        positions.append(pos + 1)  # convert to 1-based
        start = pos + 1          # +1 allows overlapping matches

    return positions

# Example usage
s = "GATATATGCATATACTT"
t = "ATAT"

result = find_motif_positions(s, t)
print(' '.join(map(str, result)))
```

```python
from Bio.SeqUtils import nt_search # nt_search() searches a nucleotide sequence for a subsequence (motif)


def find_motif_biopython(s, t):
    """
    Find all 1-based positions of motif t in sequence s using Biopython
    """
    result = nt_search(s, t)
    # ['ATAT', 1, 3, 9]

    # result[0] = motif itself
    # result[1:] = 0-based positions → we convert them to 1-based
    positions = [pos + 1 for pos in result[1:]]

    return positions

# ======================
# Example Usage (Sample Dataset)
# ======================
s = "GATATATGCATATACTT"
t = "ATAT"

positions = find_motif_biopython(s, t)
print(' '.join(map(str, positions)))
```

For this specific problem, Biopython is usually unnecessary because it's just a string-matching task.
Using Biopython doesn't provide a significant advantage here because DNA sequences are ultimately strings, and exact motif matching is already efficiently handled by Python's built-in string operations. In real bioinformatics, Biopython becomes more useful when dealing with FASTA/GenBank files, sequence alignment, translations, motifs with ambiguity, BLAST results, etc.

# Using Regular Expressions

```python
import re

def find_motif_regex_simple(s, t):
    positions = []

    # Search for all occurrences of t in s
    # (?=t) is a positive lookahead that allows overlapping matches
    for i in re.finditer(f"(?={t})", s):

        # match.start() returns the starting index (0-based)
        # Add 1 because Rosalind uses 1-based indexing
        positions.append(str(i.start() + 1))

    return positions


# Test
s = "GATATATGCATATACTT"
t = "ATAT"

print(' '.join(find_motif_regex_simple(s, t)))
```
