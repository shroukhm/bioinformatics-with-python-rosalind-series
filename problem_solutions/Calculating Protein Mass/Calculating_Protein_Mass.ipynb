```python
mass_table = {
    'A': 71.03711, 'C': 103.00919, 'D': 115.02694, 'E': 129.04259,
    'F': 147.06841, 'G': 57.02146,  'H': 137.05891, 'I': 113.08406,
    'K': 128.09496, 'L': 113.08406, 'M': 131.04049, 'N': 114.04293,
    'P': 97.05276,  'Q': 128.05858, 'R': 156.10111, 'S': 87.03203,
    'T': 101.04768, 'V': 99.06841,  'W': 186.07931, 'Y': 163.06333
}
```

## 1. Basic Programming

a simple loop to go through each amino acid and sums up their weights using a predefined dictionary.

```python
protein = "SKADYEK"
total_weight = 0

for aa in protein:
    total_weight += mass_table[aa]

print(round(total_weight, 3))
```

## Using Built-in Function

uses Python's `sum()` and a generator expression for a concise one-liner to compute the total mass.

```python
protein = "SKADYEK"
total_weight = sum(mass_table[aa] for aa in protein)

print(round(total_weight, 3))
```

## Using Biopython (ProtParam.ProteinAnalysis)

uses `Bio.SeqUtils.ProtParam` from `Biopython` to analyze protein sequences and compute their molecular weight.

```python
! pip install biopython
```

```python
from Bio.SeqUtils.ProtParam import ProteinAnalysis

protein = "SKADYEK"
analyzed = ProteinAnalysis(protein)
total_weight = analyzed.molecular_weight()

print(round(total_weight, 3))
```

🧪 Note: This Biopython method returns average molecular weight, not strictly monoisotopic. Use manual table for exact monoisotopic weight if precision is required.
