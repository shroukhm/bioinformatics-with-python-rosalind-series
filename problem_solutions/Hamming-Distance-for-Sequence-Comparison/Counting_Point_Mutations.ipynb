## 1. Basic Programming

compare each character in the two strings and counts the number of differences.

```python
s = "GAGCCTACTAACGGGAT"
t = "CATCGTAATGACGGCCT"

hamming_distance = 0

for i in range(len(s)):
    if s[i] != t[i]:
        hamming_distance += 1

print(hamming_distance)
```

## Using Python Built-in zip() Function

uses Python's `zip()` to pair elements from both strings and a generator expression to count mismatches.

```python
s = "GAGCCTACTAACGGGAT"
t = "CATCGTAATGACGGCCT"

hamming_distance = sum(1 for a, b in zip(s, t) if a != b)

print(hamming_distance)
```
